[CII] and [OI] are both cooling lines.  They represent most of the cooling channels available to the ISM.  What do we mean by cooling lines?  How can an emission line that releases photons into the ISM cool things down?

Okay.  So [CII] and [OI] represent cooling.  Now we need to think about heating.  Heating occurs when stars, shocks, or cosmic rays add energy into the ISM, making the gas in the ISM move faster and heating up larger dust grains.  We model the heating in the ISM with the total infrared luminosity (LTIR in my dataset).  What does infrared luminosity measure?  Why is it a good representation of heating?

In my data file, the [CII] Luminosity is labeled '[CII]Lum' and the [OI] Luminosity is '[OI]Lum'.  What do you expect to see if you plot the sum of these two cooling lines versus the total infrared luminosity ('LTIR' in the file).  Make a prediction before running the code below!

```python
plt.scatter(data['[CII]Lum'] + data['[OI]Lum'], data['LTIR'])
plt.show()
```
